## En un medio caracterizado por:
 $\mu =\mu_0$ y $\vec{E}=10cos(10^4t-Z)a_y$

Calcule  $\vec{H}$



Utilizando la tercera ecuacion de Maxwell o llamada de otra manera la ley de induccion Faraday:

\begin{equation}
 \nabla \times \vec{E}=-\frac{\partial\beta }{\partial t}
\end{equation}
 

Con ayuda de la ecuacion constitutiva $\beta=\mu \vec { H }$ obtengo la nueva ecuacion:


\begin{equation}
\nabla \times \vec{E}=-\mu \frac{\partial \vec{H}}{\partial t}
\end{equation}
 
 Con esta ecuacion podre obtener la intensidad de campo $\vec{H}$

Ahora integrando a ambos lados y despejando $\vec{H} $ obtengo : $\\ \\ $


\begin{equation}
\vec { H }=-\frac{1}{\mu}\int(\nabla \times E) dt\space \space (ec 1)
\end{equation}  

Para poder solucionar el integral, primero solucionare $\nabla \times E$:

\begin{equation}
\nabla \times E=
\begin{pmatrix}
\hat{x} & \hat{y} & \hat{z}\\
\frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z}\\
0 & E_y & 0
\end{pmatrix}= 
-\frac{\partial E_y}{\partial z}a_x+\frac{\partial E_y}{\partial x}a_z
\end{equation}


Reemplazando el valor de $E$ e integrando obtengo:

\begin{equation}
10sen(10^4t- Z)a_x+0
\end{equation}



Ahora voy a reemplazar el valor obtenido en la $(ec 1)$

\begin{equation}
\vec { H }=-\frac{10}{\mu}\int sen(10^4t- Z)a_x) dt
\end{equation}



Solucionando el integral obtengo que $\vec {H}$es igual a:

\begin{equation}
\vec{H}=-\frac{10}{\mu 10^4} sen(10^4t-\ Z)a_x
\end{equation}

## Un alambre recto sólido de radio $r_o$, infinitamente largo, transporta una corriente sinusoidal $i=I_m cos wt$ en estado estacionario a bajas frecuencias (sin efecto pelicular). Determine las formas de onda de la intensidad del campo magnético $H(r,t)$, la densidad de flujo magnético $B(r,t)$ y el flujo magnético $\phi(r,t)$ dentro y fuera del cable



$Solucion:$ \\

A bajas frecuencias, el efecto pelicular puede despreciarse y la corriente se distribuye uniformemente en la sección transversal del cable, como se muestra en la figura.Para determinar la intensidad del campo magnético $H(r,t)$en todas partes, se necesitan dos contornos amperianos $C_1$ y$C_2$, uno dentro del conductor para $r<=r0 $ y el otro fuera del conductor para $r>r0$

La intensidad del campo magnético en el interior del cable. La corriente en el conductor induce una intensidad de campo magnético concéntrico tanto dentro como fuera del conductor. La densidad de corriente en el interior del conductor es uniforme. Se supone que el vector de la amplitud de la densidad de corriente en el interior del conductor es paralelo al eje del conductor y viene dado por:

\begin{equation}
J_m=J_{mz}a_z
\end{equation}




FIgura1.7.jpg

$Figura 1.7$ Sección transversal de un alambre redondo y recto de longitud infinita que transporta una sinusoidal $i=I_m cos wt$ y amplitudes de densidad de corriente $J_m$, corriente encerrada $I_{m(enc)}$, e intensidad de campo magnético $H_m$ en función de la distancia radial r desde el centro del alambre a bajas frecuencias, es decir, cuando se puede despreciar el efecto piel(fasf>ro)

Consideremos un contorno radial $C_1$ en el interior del conductor. La corriente que fluye a través del área encerrada por la cáscara cilíndrica de radio $r$ a bajas frecuencias viene dada por

\begin{equation}
i_{enc}=I_{m(enc)}coswt\space\space\space\space\space\space(1.48)
\end{equation} 

donde $I_{m(enc)}$ es la amplitud de la corriente encerrada por la cáscara de radio $r$. Por lo tanto, la amplitud de la densidad de corriente en un radio $r$ es:

\begin{equation}
J_m(r)=\frac{I_{m(enc)}}{\pi r^2}\space\space\space\space For \space 0\leq r\leq r_0 \space\space\space\space(1.49)
\end{equation} 

Y la amplitud de la densidad de corriente en la superficie del cable $r=r_0$ es

\begin{equation}
J_m(r_0)=\frac{I_{m}}{\pi r_0^2} \space\space\space\space(1.50)
\end{equation} 

La densidad de corriente es uniforme a bajas frecuencias (donde se puede despreciar el efecto piel), es decir, $J_m(r)=J_m(r_0)$, dando lugar a la amplitud de la corriente encerrada:

\begin{equation}
I_{m(enc)}=I_m(\frac{\pi r^2}{\pi r_0^2})=I_m(\frac{A_r}{A_{r0}}) =I_m(\frac{r}{ r_0})^2\space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.51) \\
\end{equation} 

Donde $A_r=\pi r^2$ y $A_{r_0}= \pi r_0^2$. \\
La figura 1.7 muestra un gráfico de $I_{m(enc)}$ en función de la distancia radial desde el centro del conductor $r$. El vector de la densidad de flujo magnético es:

\begin{equation}
\textbf{H}=H_\phi=H(r)a_\phi  \space\space\space\space(1.52)
\end{equation}

De la ley de Ampere:

\begin{equation}
I_{m(enc)}=\oint_{C_1}\textbf{H}*dl=H_m(r)\oint_{C_1}dl= 2\pi rH_m(r) \space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.53)
\end{equation}


Donde $C_1=2\pi r$ para $r \leq r0$. Igualando los lados derechos de (1.51) y (1.53), se obtiene la amplitud de la intensidad del campo magnético dentro del alambre a bajas frecuencias:

\begin{equation}
H_{m}(r)=I_m(\frac{r}{r_0})\frac{1}{2\pi r}=\frac{r}{2\pi r_0^2}\space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.54) \\
\end{equation} 
La $Figura 1.7$ muestra un gráfico de la amplitud de la intensidad del campo magnético $H_m$ en función de $r$. la amplitud de la intensidad del campo magnético $H_m$ es cero en el centro del alambre porque la corriente encerrada es cero la forma de onda del campo magnético dentro del alambre a bajas frecuencias:

\begin{equation}
H(r,t)=I_m\frac{r}{2\pi r_0^2}cos \omega t\space\space\space\space For \space r\leq r_0\space\space\space\space(1.55) \\
\end{equation} 

Por lo tanto, la amplitud de la intensidad del campo magnético $H_m$ en el interior del alambre a radio $r$ está determinada únicamente por la amplitud de la corriente en el interior del radio $r$. La máxima amplitud de la intensidad del campo magnético se produce en la superficie del conductor:

\begin{equation}
H_{m(max)}=H_{m}(r_0)=\frac{I_m}{2\pi r_0}\space\space\space\space(1.56) \\
\end{equation} 

La amplitud de la densidad de flujo magnético en el interior del cable a bajas frecuencias es:

\begin{equation}
B_{m}(r)=\mu_0H_m(r)=\mu_0 I_m(\frac{r}{r_0})^2\frac{1}{2\pi r}=\mu_0I_m\frac{r}{2\pi r_0^2}\space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.57) \\
\end{equation} 

La amplitud del flujo magnético en el interior del cable a bajas frecuencias es

\begin{equation}
\phi_{m}(r)=AB_m(r)=\mu_0A_rH_m(r)=\mu_0 I_m(\frac{r(\pi r^2)}{2\pi r_0^2})=\mu_0 I_m\frac{r^3}{2 r_0^2}\space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.58) \\
\end{equation}

La forma de onda del flujo magnético es:

\begin{equation}
\phi(x,t)=\phi cos(wt)=\mu_0 I_m\frac{r^3}{2 r_0^2}cos(wt)\space\space\space\space For \space 0\leq r\leq r_0\space\space\space\space(1.59) \\
\end{equation}

La intensidad del campo magnético fuera del cable. Consideremos un contorno radial $C_2$ fuera del conductor. Toda la corriente $i=I_m cos(wt)$ está encerrada por una trayectoria de radio $r\geq r_0$. Por la ley de Ampere, la amplitud de toda la corriente $i$ es:

\begin{equation}
I_{m}=\oint_{C_2}\textbf{H}*dl=H_m(r)\oint_{C_2}dl= 2\pi rH_m(r) \space\space\space\space For \space r\geq r_0\space\space\space\space(1.60) \\
\end{equation}

Donde $C_2=2\pi r$ con $r\geq r_0$. La amplitud de la intensidad del campo magnético cercano fuera del conductor en cualquier frecuencia viene dada por la expresión:

\begin{equation}
H_{m}(r)=\frac{I_m}{2\pi r_0}\space\space\space\space For \space r\geq r_0\space\space\space\space(1.61) \\
\end{equation} 

Y la forma de onda de este campo es:

\begin{equation}
H(r,t)=\frac{I_m}{2\pi r}cos(wt)\space\space\space\space For \space r\geq r_0\space\space\space\space(1.62) \\
\end{equation} 

La amplitud de la intensidad del campo magnético aumenta linealmente con $r$ dentro del alambre desde 0 hasta $H_m(r_o)=I_m/(2\pi r_0)$ a bajas frecuencias. La amplitud de la intensidad del campo magnético es inversamente proporcional a $r$ fuera del alambre a cualquier frecuencia. 
La forma de onda de la densidad de flujo magnético es 

\begin{equation}
B(r,t)=\mu_0H(r,t)=\frac{\mu_0 I_m}{2\pi r}cos(wt)\space\space\space\space For \space  r\geq r_0\space\space\space\space(1.63) \\
\end{equation}

La forma de onda del flujo magnético encerrado por un cilindro de radio $r>r_0$ es:

\begin{equation}
\phi(x,t)=A_\omega B(r,t)=A_\omega \mu_0H(r,t)=\frac{r_0^2\mu_0 I_m}{2r}cos(wt)\space\space\space\space For \space r\geq r_0\space\space\space\space(1.64) \\
\end{equation}
